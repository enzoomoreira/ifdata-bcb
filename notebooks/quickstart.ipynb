# ifdata-bcb - Quickstart

```python
import ifdata_bcb as bcb

START = '2024-01'
END = '2025-12'

bcb.cadastro.collect(START, END)
bcb.cosif.collect(START, END)
bcb.ifdata.collect(START, END)
```

## Buscar instituicao

```python
bcb.search('Itau')
```

```python
bcb.search('Bradesco')
```

## COSIF

```python
# Listar contas
bcb.cosif.list_accounts(escopo='prudencial', limit=10)
```

```python
# Listar instituicoes
bcb.cosif.list_institutions(escopo='prudencial', start='2024-12')
```

```python
# Ler dados - escopo e obrigatorio
bcb.cosif.read(instituicao='60872504', conta=['TOTAL GERAL DO ATIVO'], start='2024-12', escopo='prudencial')
```

```python
# Multiplas instituicoes com start/end (range automatico)
bcb.cosif.read(
    instituicao=['60872504', '60746948'],
    conta=['TOTAL GERAL DO ATIVO', 'PATRIMONIO LIQUIDO'],
    start='2024-01',
    end='2024-12',
    escopo='prudencial'
)
```

## IFDATA

```python
bcb.ifdata.list_accounts(limit=10)
```

```python
# IFDATA - uma instituicao
bcb.ifdata.read(instituicao='60872504', conta=['Lucro Líquido'], start='2024-12')
```

```python
# IFDATA - multiplas instituicoes com start/end
bcb.ifdata.read(
    instituicao=['60872504', '60746948', '90400888'],
    conta=['Ativo Total', 'Lucro Líquido'],
    start='2024-01',
    end='2024-12'
)
```

```python
# Listar relatorios disponiveis
bcb.ifdata.list_reports(start='2024-12')
```

```python
# IFDATA - filtrar por relatorio (evita duplicatas quando mesma conta aparece em multiplos relatorios)
bcb.ifdata.read(
    instituicao='60872504',
    conta=['Ativo Total', 'Lucro Líquido'],
    start='2024-12',
    relatorio='Resumo'
)
```

```python
bcb.sql("""
    SELECT CNPJ_8, INSTITUICAO, VALOR
    FROM '{cache}/cosif/prudencial/*.parquet'
    WHERE DATA = 202412 AND CONTA = 'TOTAL GERAL DO ATIVO'
    ORDER BY VALOR DESC
    LIMIT 10
""")
```
